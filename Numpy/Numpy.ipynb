###  17. arr.max() ,  arr.min() ,     arr.sum() functions

```python
arr = np.random.randint(1,30,3)
print(arr)
arr = np.random.randint(1,30,(3,5))
print(arr)

print('MAX : ', arr.max())
print('MIN : ', arr.min())
print('SUM : ', arr.sum())
```

#####  We can use the axis parameter to get the max, min and sum of a particular column or row

```python
print("For Columns:")
print('MAX : ', arr.max(axis = 0))
print('MIN : ', arr.min(axis = 0))
print('SUM : ', arr.sum(axis = 0))

print("For Rows:")
print('MAX : ', arr.max(axis = 1))
print('MIN : ', arr.min(axis = 1))
print('SUM : ', arr.sum(axis = 1))
```

### 19.Sorting Function

```python
arr = np.random.randint(1,20,(3,4))
print(arr)

print("\nArray after sort :")  #default it will sort row wise
arr = np.sort(arr)
print(arr)
print("\nArray after sort col wise :")
print(np.sort(arr,axis = 0))
```

```python
arr = np.random.randint(1,20,(3,4))
print(arr)

np.sort(arr.flatten())
np.reshape(np.sort(arr.flatten()),(3,4))
np.reshape(np.sort(arr.flatten()),np.shape(arr))
np.reshape(np.sort(arr.flatten()),arr.shape)
```

###  20. Mathematical Operations

```python
np.random.seed(100)
arr = np.random.randint(1,20,(3,4))
print(arr)
print("----------------")
print(arr + 10)
print("----------------")
print(arr - 10) #Substraction
print("----------------")
print(arr * 10) #Multiplication
print("----------------")
print(arr / 10) #division
print("----------------")
print(arr ** 2) #power
print("----------------")
```

##### Matrix Multiplication

```python
np.random.seed(3)

arr1 = np.random.randint(1,10,(2,2))
arr2 = np.random.randint(1,10,(2,2))
print("arr1 :\n",arr1)
print("arr2 :\n",arr2)
print("----------------")
print(arr1.dot(arr2))
print(arr1 @ arr2)
np.dot(arr1,arr2)
```

###  25. Where function 

```python
arr = np.random.randint(1,100,10)
arr
```

```python
arr1 = np.where(arr>50 ,"High","Low")
print(arr1)

arr1 = np.where(arr%2==0, 'even',arr)
print(arr1)
arr1 = np.where(arr%2!=0, 'odd',arr1)

print(arr1)
```

###  26. Merging Arrays

##### a.Concatenate

```python
arr1 = np.random.randint(1,10,(2,4))
arr2 = np.random.randint(1,10,(2,4))

print(np.concatenate((arr1,arr2),axis=0))
print("-----------------")
print(np.concatenate((arr1,arr2),axis=1))
```

##### b.Hstack and Vstack

```python
print(np.vstack((arr1,arr2)))
print("------------------")
print(np.hstack((arr1,arr2)))
```
